```python
from keras.preprocessing.image import ImageDataGenerator
from skimage import io
import numpy as np
import os
from PIL import Image
import cv2 as cv
```

```python
none='/content/drive/MyDrive/mehboob hussain/391_Cropped_Dataset/none/'
ost='/content/drive/MyDrive/mehboob hussain/391_Cropped_Dataset/ost/'
oro='/content/drive/MyDrive/mehboob hussain/391_Cropped_Dataset/oro/'
ori='/content/drive/MyDrive/mehboob hussain/391_Cropped_Dataset/ori/'
```

```python
datagen = ImageDataGenerator(
        rotation_range=90,
        horizontal_flip=True,
        vertical_flip=True,
        height_shift_range=0.5,
        width_shift_range=0.5)
```

```python
image_directory = none
SIZE = 227
none_dataset = []

my_images = os.listdir(image_directory)
for i, image_name in enumerate(my_images):
    if (image_name.split('.')[1] == 'bmp'):
        image = io.imread(image_directory + image_name)
        image = Image.fromarray(image)
        image = image.resize((SIZE,SIZE))
        none_dataset.append(np.array(image))

none_a = np.array(none_dataset)
print(none_a.shape)
none_a= np.expand_dims(none_a, axis=3)
print(none_a.shape)
```

```python
image_directory = ori
SIZE = 227
ori_dataset = []

my_images = os.listdir(image_directory)
for i, image_name in enumerate(my_images):
    if (image_name.split('.')[1] == 'bmp'):
        image = io.imread(image_directory + image_name)
        image = Image.fromarray(image)
        image = image.resize((SIZE,SIZE))
        ori_dataset.append(np.array(image))

ori_a = np.array(ori_dataset)
print(ori_a.shape)
ori_a= np.expand_dims(ori_a, axis=3)
print(ori_a.shape)
```

```python
image_directory = ost
SIZE = 227
ost_dataset = []

my_images = os.listdir(image_directory)
for i, image_name in enumerate(my_images):
    if (image_name.split('.')[1] == 'bmp'):
        image = io.imread(image_directory + image_name)
        image = Image.fromarray(image)
        image = image.resize((SIZE,SIZE))
        ost_dataset.append(np.array(image))

ost_a = np.array(ost_dataset)
print(ost_a.shape)
ost_a= np.expand_dims(ost_a, axis=3)
print(ost_a.shape)
```

```python
i = 0
for batch in datagen.flow(none_a, batch_size=1,  
                          save_to_dir='/content/drive/MyDrive/mehboob hussain/391_Cropped_Dataset/output/none',
                          save_prefix='aug', 
                          save_format='bmp'):
    i += 1
    print(i)
    if i > 10:
      break
```
